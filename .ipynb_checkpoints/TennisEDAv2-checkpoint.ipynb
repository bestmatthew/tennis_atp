```python
from numpy import random
reversedorder = ['carpet', 'clay', 'grass', 'hard', 'best_of', 'loser_ht', 'loser_age','loser_rank','L2', 'R2', 'U2', 'winner_ht', 'winner_age','winner_rank', '1L', '1R', '1U']
```

```python
reorders = {
    'winner_ht': 'a_ht',
    'winner_age': 'a_age',
    'winner_rank': 'a_rank',
    '1L': 'aL',
    '1R': 'aR',
    '1U': 'aU',
    'loser_ht': 'b_ht',
    'loser_age': 'b_age',
    'loser_rank': 'b_rank',
    'L2': 'bL',
    'R2': 'bR',
    'U2': 'bU'
    }



def flipornot(row):

    flip = np.random.choice([0,1], p=[0.5,0.5])
    
    if flip == 0: #player info does not shuffle, player A is winner (default)
        row['result'] = 0
    else:   #winner info goes into player B spot
        row = row.reindex(reversedorder)
        row['result'] = 1
    #rename here
    
    return row.rename(dict(zip(row.index[-len(keys):], keys)))


viewnew = b2010.apply(lambda x: flipornot(x), axis=1)
```
